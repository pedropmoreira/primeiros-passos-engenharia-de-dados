PAGINANDO OS REPOSITORIOS:

```python
api_base_url = 'https://api.github.com'
owner = 'amzn' # username de quem vamos extrair os dados
url = f'{api_base_url}/users/{owner}/repos'

url
```

```python
repos_list = []
for page_num in range(1, 6):
    try:
        url_page = f'{url}?page={page_num}'
        response = requests.get(url_page, headers=headers)
        repos_list.append(response.json())
    except:
        repos_list.append(None)
```

```python
repos_list
```

```python
len(repos_list)# as cinco paginas de repositorio
```

```python
len(repos_list[0])# todos repositorios da primeira pagina
```

NOMES DOS REPOSITORIOS:

```python
repos_list
```

```python
# acessando nomes dos repos
repos_list[0][2]['name']
```

```python
repos_name = []
for page in repos_list:
    for repo in page:
        repos_name.append(repo['name'])
```

```python
# checando se o numero de repositorios na lista repos_name √© igual ao total de repositorios do github
len(repos_name)
```

SELECIONANDO AS LINGUAGENS:

```python
repos_language = []
for page in repos_list:
    for repo in page:
        repos_language.append(repo['language'])
```
